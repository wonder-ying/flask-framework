```python
import requests
import pandas as pd
import datetime

from bokeh.embed import server_document, components 
from bokeh.layouts import column
from bokeh.models import ColumnDataSource, HoverTool, TextInput, CustomJS
from bokeh.io import curdoc
from bokeh.plotting import figure, output_file, show
from bokeh.server.server import Server
from bokeh.themes import Theme
```

```python
API_URL = "https://www.alphavantage.co/query"
data = { "function": "TIME_SERIES_DAILY_ADJUSTED","symbol": "AAPL","apikey": "UBSVT7NO385KCUNM","outputsize": "full","datatype": "json"}
r = requests.get(API_URL, data)
r_json = r.json()
```

```python
r_json['Time Series (Daily)']
```

```python
df = pd.DataFrame.from_dict(r_json['Time Series (Daily)'], orient='index').sort_index()
df.reset_index(inplace = True)
df
```

```python
df = df.rename(columns={ 'index': 'Date','1. open': 'Open', '2. high': 'High', '3. low': 'Low', '4. close': 'Close', '5. adjusted close': 'Adjusted Close', '6. volume': 'Volume', '7. dividend amount': 'Dividend Amount', '8. split coefficient': 'Split Coefficient'})
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df1 = df.loc[df.Date>='2017-01-01', ['Date', 'Open', 'Close', 'Adjusted Close']]
df1.info()
```

```python
a = figure(x_axis_type="datetime", plot_width=1200, plot_height=600, title='Stock Price Tracking', toolbar_location="below")
a.line(df1['Date'], df1['Open'], line_width=2, line_color='blue', legend_label='Open')
show(a)
```

```python
# Plot the graph in Bokeh
p = figure(x_axis_type="datetime", plot_width=1200, plot_height=600, title='Stock Price Tracking', toolbar_location="below")
p.outline_line_color = "black"
p.xgrid.grid_line_color=None
p.ygrid.grid_line_alpha=0.5
p.title.text = 'Daily Stock Price Tracking of %s' % 'FB'.upper()
p.xaxis.axis_label = "Date and Year"
p.yaxis.axis_label = "Stock Price"
p.title.text_font_size = "20px"
p.line(df1['Date'], df1['Open'], line_width=2, line_color='blue', legend_label='Open')
#p.line(df1['Date'], df1['Close'], line_width=2, line_color='Green', legend_label='Close')
p.add_tools(HoverTool(
  tooltips=[
     ( 'Date',   '@x{%F}'     ),
     ( 'Price',  '$@y{%0.2f}' )
  ],

  formatters={
     '@x'      : 'datetime',
     '@y'     : 'printf',
  },

  mode='mouse'
))
show(p)
```
